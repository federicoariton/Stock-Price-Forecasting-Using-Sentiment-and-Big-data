# Clean the Tweet Data

```python
import pandas as pd
import re

# Load the tweet dataset from your local folder (downloaded from HDFS)
tweets_df = pd.read_csv("../data/stocktweet/stocktweet.csv")

# Drop missing values in critical columns
tweets_df.dropna(subset=["date", "ticker", "tweet"], inplace=True)

# Standardize ticker format (uppercase, remove $)
tweets_df["ticker"] = tweets_df["ticker"].str.upper().str.replace("$", "", regex=False)

# Format the date
tweets_df["date"] = pd.to_datetime(tweets_df["date"], dayfirst=True).dt.strftime("%Y-%m-%d")

```

## Analyzing the companies with the most tweet volume

```python
# Group by ticker and count tweets
mention_counts = tweets_df.groupby("ticker").size().reset_index(name="tweet_volume")

# Sort in descending order
mention_counts_sorted = mention_counts.sort_values(by="tweet_volume", ascending=False)

# View top 15
mention_counts_sorted.head(15)

```

```python
import matplotlib.pyplot as plt

top_n = 15
mention_counts_sorted.head(top_n).plot(
    kind="bar", x="ticker", y="tweet_volume", figsize=(12,6),
    title=f"Top {top_n} Most Mentioned Tickers on Twitter",
    legend=False, grid=True
)
plt.ylabel("Tweet Volume")
plt.xlabel("Ticker")
plt.tight_layout()
plt.show()
```

```python
# Cleaning tweet text
tweets_df["tweet_clean"] = tweets_df["tweet"].str.lower().apply(lambda x: re.sub(r"[^a-zA-Z\s]", "", x))
```
