```python
import pandas as pd
from pymongo import MongoClient

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Access the database and collection
db = client["stock_data"]
collection = db["sentiment"]

# Pull all documents from the collection
documents = list(collection.find())

# Convert to DataFrame
df = pd.DataFrame(documents)

# Drop MongoDB's internal _id field
df.drop(columns=['_id'], inplace=True, errors='ignore')


df.head()
```

```python
# Access the ARIMA/SARIMAX results
db_arima = client["stock_forecast"]
arima_col = db_arima["results"]

# Load and convert to DataFrame
arima_docs = list(arima_col.find())
df_arima = pd.DataFrame(arima_docs).drop(columns=["_id"], errors="ignore")

# Preview
df_arima.head()
```

```python
def flatten_forecast_df(df_raw):
    rows = []
    for _, row in df_raw.iterrows():
        for i in range(len(row['dates'])):
            rows.append({
                'ticker': row['ticker'],
                'model': row['model'],
                'date': pd.to_datetime(row['dates'][i]),
                'forecast_close': row['forecast_real'][i] if i < len(row['forecast_real']) else None,
                'true_close': row['true_real'][i] if i < len(row['true_real']) else None,
                'rmse': row['rmse'],
                'forecast_horizon': row['horizon'],
                'mode': row['mode']
            })
    return pd.DataFrame(rows)

# Flatten the forecast output
df_forecast_flat = flatten_forecast_df(df_arima)
```

```python
df['date'] = pd.to_datetime(df['date'])
df_hist = df[['ticker', 'date', 'Close']].copy()
df_hist.rename(columns={'Close': 'true_close'}, inplace=True)
df_hist['forecast_close'] = None
df_hist['model'] = None
df_hist['forecast_horizon'] = None
df_hist['rmse'] = None
df_hist['mode'] = 'actual'
```

```python
df_combined = pd.concat([df_hist, df_forecast_flat], ignore_index=True)
df_combined = df_combined.sort_values(by=['ticker', 'date', 'mode']).reset_index(drop=True)
```

```python
df_combined
```

```python
import matplotlib.pyplot as plt

# Select a ticker to plot
ticker = "AAPL"
df_plot = df_combined[df_combined['ticker'] == ticker].copy()

# Clean list values if they exist (just in case)
df_plot['forecast_close'] = df_plot['forecast_close'].apply(lambda x: x[0] if isinstance(x, list) else x)
df_plot['true_close'] = df_plot['true_close'].apply(lambda x: x[0] if isinstance(x, list) else x)
df_plot['date'] = pd.to_datetime(df_plot['date'])

# Plot setup
plt.figure(figsize=(14, 6))

# Actual values
actual = df_plot[df_plot['mode'] == 'actual']
plt.plot(actual['date'].values, actual['true_close'].values, label='Actual', color='black')

# Evaluation forecast
eval_forecast = df_plot[df_plot['mode'] == 'evaluation']
plt.plot(eval_forecast['date'].values, eval_forecast['forecast_close'].values, label='Forecast (Eval)', linestyle='--', marker='o')

# Future forecast
future_forecast = df_plot[df_plot['mode'] == 'future']
plt.plot(future_forecast['date'].values, future_forecast['forecast_close'].values, label='Forecast (Future)', linestyle='--', marker='x')

# Optional: Mark forecast start
last_actual_date = actual['date'].max()
plt.axvline(x=last_actual_date, color='gray', linestyle='--', label='Forecast Start')

# Final plot settings
plt.title(f"{ticker} - Actual vs Forecasted Close Prices")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Choose a ticker (e.g., AAPL)
ticker = "AAPL"

# Filter only relevant rows
df_plot = df_combined[
    (df_combined['ticker'] == ticker) &
    (df_combined['mode'].isin(['actual', 'future']))
].copy()

# Fix any list-type values
df_plot['forecast_close'] = df_plot['forecast_close'].apply(lambda x: x[0] if isinstance(x, list) else x)
df_plot['true_close'] = df_plot['true_close'].apply(lambda x: x[0] if isinstance(x, list) else x)
df_plot['date'] = pd.to_datetime(df_plot['date'])

# Separate actual and future forecast data
df_actual = df_plot[df_plot['mode'] == 'actual']
df_future = df_plot[df_plot['mode'] == 'future']

# Plot
plt.figure(figsize=(14, 6))
plt.plot(df_actual['date'], df_actual['true_close'], label='Historical Close', color='black')
plt.plot(df_future['date'], df_future['forecast_close'], label='Forecasted Close (Future)', linestyle='--', marker='o', color='orange')

# Mark the forecast start
if not df_actual.empty:
    last_actual_date = df_actual['date'].max()
    plt.axvline(x=last_actual_date, linestyle='--', color='gray', label='Forecast Start')

# Decorate
plt.title(f"{ticker} - Historical & Future Forecasted Close Prices")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
